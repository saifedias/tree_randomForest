```python
"""
standards
"""
#import numpy as np
#import pandas as pd
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']

"""
own
"""
from my_functions import plot_train_test

"""
scikit-learn
"""
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data = load_wine(as_frame=True)
X = data.data
y = data.target
```

```python
X.head()
```

```python
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

plot_train_test(X_train, X_test, y_train, y_test)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True, dpi=100)
for header, ax, y_i, sct in zip(["True", "Predicted"], axs, [y_test, clf.predict(X_test)], ["Training", "Test"]):
    ax.set_title(header)
    ax.scatter(X_test["alcohol"], X_test["malic_acid"], c=y_i, s=75, label=accuracy_score(y_test, y_i).round(3))
    ax.grid(color="k", alpha=0.1, zorder=0)
    ax.legend(loc="best")
plt.show()
```

```python
clf = RandomForestClassifier(n_estimators=200)
clf.fit(X_train, y_train)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True, dpi=100)
for header, ax, y_i, sct in zip(["True", "Predicted"], axs, [y_test, clf.predict(X_test)], ["Training", "Test"]):
    ax.set_title(header)
    ax.scatter(X_test["alcohol"], X_test["malic_acid"], c=y_i, s=75, label=accuracy_score(y_test, y_i).round(3))
    ax.grid(color="k", alpha=0.1, zorder=0)
    ax.legend(loc="best")
plt.show()
```
