```python
"""
standard
"""
import numpy as np
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']

"""
own
"""
from my_functions import plot_train_test, plot_true_predicted, plot_validation_curve

"""
scikit-learn
"""
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold
```

---
## load data
using [sklearn's wine dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_wine.html)

```python
data = load_wine(as_frame=True)
X = data.data
y = data.target
```

```python
X.head()
```

```python
y.unique()
```

---
## separate train and test data

using [scikit-learn's train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

plot_train_test(X_train, X_test, y_train, y_test)
```

---
## Decision-Tree

using [scikit-learn's DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier)

here: I'm using DT straight out of box ***without hyperparameter tuning***

&rarr; unpruned DT

```python
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt.fit(X_train, y_train)
```

```python
clf_dt.get_depth()
```

```python
plot_true_predicted(X_test, y_test, clf_dt)
```

---
## Random-Forest

using [scikit-learn's RandomForestClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html#sklearn.ensemble.RandomForestClassifier)

here: I'm using RF straight out of box (except for the number of trees) ***without hyperparameter tuning***

```python
clf_rf = RandomForestClassifier(n_estimators=200, random_state=1)
clf_rf.fit(X_train, y_train)
```

```python
plot_true_predicted(X_test, y_test, clf_rf)
```

# tuning Hyperparameter

### DT

```python
param_grid = {"max_depth": [2, 3, 4, 5, 6, 7, 8]}

clf = GridSearchCV(DecisionTreeClassifier(),
                   param_grid,
                   n_jobs=12,
                   cv=RepeatedStratifiedKFold(n_repeats=15, n_splits=10)
                  )

clf.fit(X_train, y_train)
print(clf.best_params_)

cv_results = pd.DataFrame(clf.cv_results_)
cv_results.head()
```

```python
plot_validation_curve(cv_results)
```

---
### RF

```python
param_grid = {"max_depth": [2, 3, 4, 5, 6, 7, 8, 9, 10]}

clf = GridSearchCV(RandomForestClassifier(),
                   param_grid,
                   n_jobs=12,
                   cv=RepeatedStratifiedKFold(n_repeats=15, n_splits=10)
                  )

clf.fit(X_train, y_train)
print(clf.best_params_)

cv_results = pd.DataFrame(clf.cv_results_)
cv_results.head()
```

```python
plot_validation_curve(cv_results)
```

```python
param_grid = {"n_estimators": [5, 10, 20, 40, 80, 160, 320, 640]}

clf = GridSearchCV(RandomForestClassifier(),
                   param_grid,
                   n_jobs=12,
                   cv=RepeatedStratifiedKFold(n_repeats=15, n_splits=10)
                  )

clf.fit(X_train, y_train)
print(clf.best_params_)

cv_results = pd.DataFrame(clf.cv_results_)
cv_results.head()
```

```python
plot_validation_curve(cv_results)
```

```python
param_grid = {"max_features": np.arange(1, len(X.columns)+1, 1)}

clf = GridSearchCV(RandomForestClassifier(n_estimators=200),
                   param_grid,
                   n_jobs=12,
                   cv=RepeatedStratifiedKFold(n_repeats=15, n_splits=10)
                  )

clf.fit(X_train, y_train)
print(clf.best_params_)

cv_results = pd.DataFrame(clf.cv_results_)
cv_results.head()
```

```python
plot_validation_curve(cv_results)
```
